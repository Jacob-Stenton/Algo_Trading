# SMA Crossover Algorithmic Trading Strategy

This is an area to learn some basics of algo trading and backtesting based on a simple smoothed moving average (SMA) crossover strategy.

## Import Libraries

```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

## Import Data


### Downloading forex data

```python
ticker = 'GBPUSD=X'
data = yf.download(ticker, period='3mo', interval='1h') # Previous month and the 1 hour time frame - yf.download --> pandas dataframe
```

```python
data
```

### Plot of forex data

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(data['Close'])
plt.show()
```
